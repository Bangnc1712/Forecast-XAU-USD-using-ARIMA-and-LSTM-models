##Tổng hợp các mô hình SARIMA

(P,D,Q): (5,1,5); (5,1,11); (11,1,5) và (11,1,11)

(p,d,q): (5,1,5); (5,1,11); (11,1,5) và (11,1,11)

m = 12 (chu kì 12 tháng)

Mô hình SARIMA được ký hiệu là

SARIMA(p, d, q)(P, D, Q)m

, trong đó:

p, d, q: Là các tham số của thành phần ARIMA (phi mùa vụ):

p: Bậc của thành phần tự hồi quy (AR).

d: Bậc của thành phần tích phân (I).

q: Bậc của thành phần trung bình trượt (MA).

P, D, Q: Là các tham số của thành phần mùa vụ:

P: Bậc của thành phần tự hồi quy mùa vụ.

D: Bậc của thành phần tích phân mùa vụ.

Q: Bậc của thành phần trung bình trượt mùa vụ.

m: Chu kỳ mùa vụ (ví dụ: 12 cho dữ liệu hàng tháng, 4 cho dữ liệu hàng quý).

=> các mô hình SARIMA ta có:

1. SARIMA(5,1,5)(5,1,5)[12]
2. SARIMA(5,1,5)(5,1,11)[12]
3. SARIMA(5,1,5)(11,1,5)[12]

##Đồ thị phần dư

```r
# Đồ thị phần dư
par(mfrow = c(1, 2))
plot(residuals(modelA), main = "Residuals of modelA", ylab = "Residuals")
plot(residuals(modelAuto), main = "Residuals of modelAuto", ylab = "Residuals")
```

##Kiểm tra Ljung-Box cho phần dư


```r
Box.test(residuals(modelA), lag = 12, type = "Ljung-Box")
Box.test(residuals(modelAuto), lag = 12, type = "Ljung-Box")
```

```r
plot(modelA)
```

```r
plot(modelAuto)
```

```r
eA <- arima.errors(modelA)
shapiro.test(eA)

eAuto <- arima.errors(modelA)
shapiro.test(eAuto)
```
